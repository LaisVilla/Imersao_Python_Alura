```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Cdigo criado com uma API YF: Que contm dados de aes de bolsas abertas.
# Api: Uma interface que conecta o cdigo a alguma base de dados.

dados = yf.download('PETR4.SA', start= '2023-01-01', end= '2023-12-31')
dados
```

```python
# Renomendo colunas do df todas de uma nica vez.
dados.columns = ['Abertura', 'Maximo', 'Minimo', 'Fechamento', 'Fech_Ajuste', 'Volume']
```

```python
#renomeando apenas o nome da coluna indice.
dados = dados.rename_axis('Data')
dados
```

```python
#criando o primeiro grafico usando a coluna fechamento usando a biblioteca do pd determinado o tamanho.
dados['Fechamento'].plot(figsize= (10,6))
plt.title('Variao do preo por data', fontsize=16)
plt.legend(['Fechamento'])
```

```python
df = dados.head(60).copy()
df

# Convertendo o indice em uma coluna de data
df['Data'] = df.index

# Convertendo as datas para o formato nmerico de matplotlib
# Isso  necessrio para o matplotlib possa plotar as datas corretamente no grfico
# Apply  uma funo que recebe um outra funo para voc aplicar na sua srie de dados toda.
# MDATES  um metdo do Matplot lib para a manipulao de datas.
df['Data'] = df['Data'].apply(mdates.date2num)

df
```

```python
# Fig e ax so resultados da funo subplots, cria layouts/ tela inicial.
fig, ax = plt.subplots(figsize=(15,8))
# Vamos definir a largura dos candles no grfico
width = 0.7

# Len ir contar o tamanho do df, range est contando 0,1,2,3... a 59.
for i in range(len(df)):
    # Determinar a cor do candle
    # Se o preo de fechamento for maior que o da abertura, o candle  verde (a ao valorizou nesse dia).
    # Se for menor, o candle  vermelho (a ao desvalorizou).
    if df['Fechamento'].iloc[i] > df['Abertura'].iloc[i]:
      color = 'green'
    else:
      color = 'red'

  # Desenhando a linha vertical do candle (mecha).
  # Essa linha mostra os preos mximo (topo da linha) e mnimo (base da linha) do dia.
  # Usamos 'ax.plot' para desenhar uma linha vertical.
  # [df['Data'].iloc[i], df['Data'].iloc[i]] define o ponto x da linha (a data), e [df['Mnimo'].iloc[i], df['Mximo'].iloc[i]] define a
    ax.plot([df['Data'].iloc[i], df['Data'].iloc[i]],
          [df['Minimo'].iloc[i], df['Maximo'].iloc[i]],
          color=color,
          linewidth=1)
    # add_patch  o metdo para adicionar objetos no matplotlib

    ax.add_patch(plt.Rectangle((df['Data'].iloc[i] - width/2, min(df['Abertura'].iloc[i], df['Fechamento'].iloc[i])),
                               width,
                               # abs  uma abreviao para absoluto 'ser sempre um nmero absoluto, positivo'.
                               abs(df['Fechamento'].iloc[i] - df['Abertura'].iloc[i]),
                               facecolor=color))

df['MA7'] = df['Fechamento'].rolling(window=7).mean()
df['MA14'] = df['Fechamento'].rolling(window=14).mean()

# Mdias Mveis -> Indica tendncia, pega um periodo longo de tempo e voc tira uma mdia daquele periodo e gera um ponto aqui,
# depois retira novamente um longo periodo de uma janela igual, ento vce faz uma janela de mdias que vai te gerando os pontos. Exemplo de mdias moveis de mortos na pandemia.
# Plotando as mdias mveis
ax.plot(df['Data'], df['MA7'], color='orange', label='Mdia Mvel 7 dias') # Mdia de 7 dias
ax.plot(df['Data'], df['MA14'], color='yellow', label='Mdia Mvel 14 dias') # Mdia de 14 dias
# Adicionando legendas para as Mdias Moveis
ax.legend()

# Formatando o eixo x para mostrar as datas
# Configuramos o formato da data e a rotao para melhor legibilidade
ax.xaxis_date() #O mtodo xaxis_date()  usado para dizer ao Matplotlib que as datas esto sendo usadas no eixo x
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

# Adicionando ttulo e rtulos para os eixos x e y
plt.title("Grfico de Candlestick - PETR4.SA com matplotlib")
plt.xlabel("Data")
plt.ylabel("Preo")

# Adicionando uma grade para facilitar a visualizao dos valores
plt.grid(True)

# Exibindo o grfico
plt.show()
```

```python
# Criando subplots
'''
"Primeiro, criamos uma figura que conter nossos grficos usando make_subplots.
Isso nos permite ter mltiplos grficos em uma nica visualizao.
Aqui, teremos dois subplots: um para o grfico de candlestick e outro para o volume de transaes."

'''
fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                    vertical_spacing=0.1,
                    subplot_titles=('Candlesticks', 'Volume Transacionado'),
                    row_width=[0.2, 0.7])

'''
"No grfico de candlestick, cada candle representa um dia de negociao,
mostrando o preo de abertura, fechamento, mximo e mnimo. Vamos adicionar este grfico  nossa figura."
'''
# Adicionando o grfico de candlestick
fig.add_trace(go.Candlestick(x=df.index,
                             open=df['Abertura'],
                             high=df['Maximo'],
                             low=df['Minimo'],
                             close=df['Fechamento'],
                             name='Candlestick'),
                             row=1, col=1)

# Adicionando as mdias mveis
# Adicionamos tambm mdias mveis ao mesmo subplot para anlise de tendncias
fig.add_trace(go.Scatter(x=df.index,
                         y=df['MA7'],
                         mode='lines',
                         name='MA7 - Mdia Mvel 7 Dias'),
                         row=1, col=1)

fig.add_trace(go.Scatter(x=df.index,
                         y=df['MA14'],
                         mode='lines',
                         name='MA14 - Mdia Mvel 14 Dias'),
                         row=1, col=1)

# Adicionando o grfico de barras para o volume
# Em seguida, criamos um grfico de barras para o volume de transaes, que nos d uma ideia da atividade de negociao naquele dia
fig.add_trace(go.Bar(x=df.index,
                     y=df['Volume'],
                     name='Volume'),
                     row=2, col=1)

# Atualizando layout
#Finalmente, configuramos o layout da figura, ajustando ttulos, formatos de eixo e outras configuraes para tornar o grfico claro e legvel.
fig.update_layout(yaxis_title='Preo',
                  xaxis_rangeslider_visible=False,  # Desativa o range slider
                  width=1100, height=600)

# Mostrando o grfico
fig.show()
```

```python
dados_vale = yf.download('VALE3.SA', start='2019-01-01', end='2019-12-31')
```

```python
# Calcular as mdias mveis exponenciais (EMA)
dados_vale['EMA_20'] = dados_vale['Close'].ewm(span=20, adjust=False).mean()
dados_vale['EMA_50'] = dados_vale['Close'].ewm(span=50, adjust=False).mean()

# Configurar o layout do grfico
layout = mpf.make_mpf_style(base_mpf_style='mike', y_on_right=True)

# Criar grfico de candlestick com volume e mdias mveis exponenciais no lado esquerdo
mpf.plot(dados_vale, type='candle',figsize = (18,7), volume=True, title='Grfico de Candlestick da Vale em 2019',
         mav=(20, 40), ylabel='Preo', ylabel_lower='Volume', style=layout)
```
