```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Acoes_pura.xlsx", sheet_name="Principal")
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel("/content/Acoes_pura.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/Acoes_pura.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_principal_chatgpt= pd.read_excel("/content/Acoes_pura.xlsx", sheet_name="Chatgpt")
df_principal_chatgpt
```

```python
df_principal = df_principal[['Ativo','Data','Último (R$)','Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct']/100
df_principal['valor_inicial'] = df_principal['valor_final']/(df_principal['Var_pct']+ 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_principal_chatgpt, left_on='Nome', right_on='Nome da empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da empresa'])
df_principal = df_principal.rename(columns={'Idade (anos)':'Idade'}).copy()
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['Idade'].apply(lambda x: 'Empresas Jovens' if x < 20 else ('Empresas em Crescimento' if x >= 20 and x < 50 else ('Empresas Estabelecidas' if x >= 50 and x < 100 else 'Empresas Centenárias')))
df_principal
```

```python

# Calcular a maior variação
maior_variacao = df_principal['Variacao_rs'].max()

# Calcular a menor variação
menor_variacao = df_principal['Variacao_rs'].min()

# Calcular a média das variações
media_variacoes = df_principal['Variacao_rs'].mean()

# Calcular a média das variações para os casos em que o resultado é "Subiu"
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calcular a média das variações para os casos em que o resultado é "Desceu"
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Imprimir os resultados
print("Maior\t\tR$ {:,.2f}".format(maior_variacao))
print("Menor\t\tR$ {:,.2f}".format(menor_variacao))
print("Média\t\tR$ {:,.2f}".format(media_variacoes))
print("Média de quem subiu\tR$ {:,.2f}".format(media_subiu))
print("Média de quem desceu\tR$ {:,.2f}".format(media_desceu))
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text= 'Variacao_rs', title= 'Variação Reais por Resultado')
# Transformar formatação dos números do gráfico de barras;
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.show()
```

# Gráfico de pizza no df_análise_segmentos com a mesma biblioteca Potly;


```python
# Criar o gráfico de pizza
fig = px.pie(df_analise_segmento, values='Variacao_rs', names='Segmento',
             title='Variação Reais por Segmento')

# Exibir o gráfico
fig.show()
```

```python
df_analise_idade = df_principal.groupby('Cat_idade')['Variacao_rs'].sum().reset_index()
df_analise_idade
```

```python
# Código criado com ajuda do chatgpt
# Criando o gráfico de barras com Plotly Express
fig = px.bar(df_analise_idade, x='Cat_idade', y='Variacao_rs', title='Variação em RS por Categoria de Idade')

# Atualizando o layout do gráfico para melhorar a legibilidade dos rótulos do eixo x
fig.update_layout(xaxis=dict(tickangle=45))

# Exibindo o gráfico
fig.show()
```

```python
# Contando o número de ocorrências de cada valor na coluna 'Cat_idade'
contagem_cat_idade = df_principal['Cat_idade'].value_counts().reset_index()

# Renomeando as colunas resultantes
contagem_cat_idade.columns = ['Cat_idade', 'Número de Empresas']


display(contagem_cat_idade)
```
